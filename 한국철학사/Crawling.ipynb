```python
try:
    while True:
        
        now = dt.now()
        KST = now + datetime.timedelta(hours=9)
        date = KST.strftime('%Y-%m-%d %H:%M:%S')

        news_url = 'https://news.daum.net/'
        req = requests.get(news_url)
        soup = BeautifulSoup(req.text, 'html.parser')

        print()
        print('변수 초기화')

        news_dict = {}
        press = []
        idx = 0
        table = None
        li_list = None
        area_list = None
        a_list = None
        pressCover_list = None
        press_list = None


        print()
        print('크롤링 중...')


        table = soup.find('ul',{'class' : 'list_issue'})
        li_list = table.find_all('li', {'data-tiara-layer': re.compile('cluster*')})
        area_list = [li.find('div', {'class' : 'item_issue'}) for li in li_list]
        a_list = [area.find('a', {'class' : 'link_txt'}) for area in area_list]
        pressCover_list = [area.find('div', {'class' : 'cont_thumb'}) for area in area_list]
        press_list = [press.find('span', {'class' : 'info_thumb'}) for press in pressCover_list]

        for i in press_list[:min(len(press_list), news_num-idx)]:
            press.append(i.text)


        for n in a_list[:min(len(a_list), news_num-idx)]:
            news_dict[idx] = {        
                'time' : date,
                              'title' : n.text,
                              'url' : n.get('href'),
                              'press' : press[idx],
                             }
            idx += 1


        print('크롤링 완료')
        print('데이터 프레임 변환')
        news_df = DataFrame(news_dict).T

        folder_path = os.getcwd()
        csv_file_name = '다음뉴스_{}.csv'.format(date)

        news_df.to_csv(csv_file_name,index=False, encoding="utf-8-sig")
        # news_df.to_csv(csv_file_name,mode ='a', header = False, index = False)

        # with open(csv_file_name, mode = 'a') as f:
        #     news_df.to_csv(f, header=f.tell()==0,index = False)


        print('새로운 csv 저장 완료 | 경로 : {}\\{}'.format(folder_path, csv_file_name))

        print('30분 동안 멈춥니다..')
        time.sleep(1800)
        print('30분이 지났기에 다시 크롤링을 시작합니다.')
except KeyboardInterrupt:
    #releas 코드
    pass

```
