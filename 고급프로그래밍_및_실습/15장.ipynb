```python
#2번 각 사람의 BMI를 구하여 열에 추가하세요.
import csv
from google.colab import files
import pandas as pd
# file_uploaded = files.upload()

list1 = pd.read_csv('/content/studentlist.csv')
#bmi = weight/(height**2)

list1['BMI'] = list1['weight']/(list1['height']/100)**2

list1
```

```python
#3번 두 개의 꺾은선그래프그리세요.
#-1번: x축은 이름 y축은 몸무게를 나타내는 그래프 1개
#-2번: x축은 이름 y축은 몸무게를 나타내는 그래프 1개
import csv
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

ax = plt.gca()
list1 = pd.read_csv('/content/studentlist.csv')

list1.plot(kind='line', x='name', y='weight', ax=ax)
list1.plot(kind='line', x='name', y='height', ax=ax)

plt.show()
```

```python
#4번 titanic data에서 class의 Fare 평균을 구하세요.
import csv
from google.colab import files
import pandas as pd
# file_uploaded = files.upload()

import seaborn as sns
import pandas as pd

titanic= sns.load_dataset("titanic")
titanic.groupby(['class'])['fare'].mean()

```

```python
#5번 titanic 데이터에서 alive == yes, age는 25이상, pclass는 1보다 큰 데이터만 남겨 아래와 같이 나타내세요
import seaborn as sns
import pandas as pd
titanic= sns.load_dataset("titanic")

titanic = titanic[titanic["alive"]== "yes"]
titanic = titanic[titanic["age"]>=25]
titanic = titanic[titanic["pclass"]>1]

titanic
```

```python
# 6번 studentlist의 혈액형 별 도수를 bar 차트로 그리시오
import matplotlib.pyplot as plt

list1 = pd.read_csv('/content/studentlist.csv')
list_count = list1['bloodtype'].value_counts()
list_count.to_frame(name='bloodtype')
list_count.plot(kind ='bar')
plt.show()
```
