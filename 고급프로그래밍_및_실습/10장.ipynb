```python
#2번
#사용자로 부터 비밀번호를 입력 받고, 정규식을 사용하여 다음의 조건들을 만족하는지 검증하는 프로그램을 작성하세요.
#1. 8글자이상 13글자 미만
#2. 대문자 한개 이상 포함
#3. 특수문자 포함하지 않아야 함.
#4. 문자중 x가 들어가면 안 된다.
#예시: whxgjs3131(x, 4번조건 위배)

import re

password = input("패스워드를 입력하세요: ")
flag = 0

if (len(password) <8 or len(password) >=13):
  flag = -1
  print(1)
elif not re.search("[A-Z]", password):
  flag = -1
  print(2)
elif re.search('[_, @, $]',password) :
  flag = -1
  print(3)
elif re.search("x", password):
  flag = -1
  print(4)
else:
  flag = 0
  print("유효한 패스워드입니다.")

if flag == -1:
  print("유효한 패스워드가 아닙니다.")
```
