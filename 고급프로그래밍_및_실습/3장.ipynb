```
#1번 : 95점이상이면 A+ 90점이상이면 A 80점이상이면 B를 출력하는 프로그램을 만들어라

score = int(input('점수를 입력하세요'))
grade = 'O'

if score >= 95 :
  grade = 'A+'
elif score >= 90 :
  grade = 'A'
elif score >=80 :
  grade ='B'
else : 
  grade ='c'

print(grade)

```

```
#2번: 다음은 당신의 프로그래밍1 학점에 대한 개발자역량을 평가하는 가상 식이다

#식 = (4.5-프로그래밍1성적)/10 (0이 최고프로그래머, 0.45 가장 최악의 점수 다른 길 알아보기!)
#프로그래밍 1성적을 입력받고 0.15미만이면 '프로그래머한번해보자!', 0.15이상이면 '다시 한번 생각해볼까?' 를 출력하라


score = float(input('프로그래밍1성적을 입력하세요'))

a = (4.5 - score)/10

if a < 0.15 :
  print('프로그래머한번해보자!')
else : 
  print('다시 한 번 생각해볼까?')

```

```
#3번:세 개의 숫자를 입력받았을때 삼각형의 결정조건에 해당하는지 판별하는 프로그램을 작성해보자

a = int(input('첫 번째 숫자를 입력하세요'))
b = int(input('두 번째 숫자를 입력하세요'))
c = int(input('세 번째 숫자를 입력하세요'))

if (a+b) > c and (a+c) > b and (b+c) > a :
  print('올바른 삼각형 입니다.')
else :
  print('틀려먹은 삼각형 입니다.')

```

```
#4번: 한 개의 3자리숫자를 입력받아서 5의배수인지 아니면 아무것도 아닌지 판별하는 프로그램을 만들어라
#4-2번: 한 개의 2자리숫자를 입력받아서 2의배수이거나 3의 배수이면서 6의 배수는 아닌 숫자면 OK를 출력 아니면 NO를 출력하는 프로그램을 만들어라.

a=int(input("3자리 숫자를 입력해주세요"))

#4
if (100 <= a and a <=999) :
  if a % 5 == 0 :
    print('5의 배수입니다')
  else :
    print('5의 배수가 아닙니다.')
else :
  print('입력 오류')



#4-2
b = int(input('2자리 숫자를 입력해주세요'))

if (10 <= b and b <=99 ) :  
  if (b % 2 == 0 or b % 3 == 0) and b % 6 !=0 :
    print('OK')
  else : 
    print('NO')
else : 
  print('입력 오류')


```

```
#5번: 나이에서 20~22을 20대 초반 23~26을 중반 27~29를 후반이라고 한다. 나이를 입력했을 때 어디 영역인지 출력하는 함수를 만들어라.

age = int(input('20대의 정확한 나이를 입력해주세요'))

if (age >=20 and age<=22) :
  print('20대 초반입니다.')
elif (age>=23 and age<=26) :
  print('20대 중반입니다.')
elif (age>=27 and age <=29) :
  print('20대 후반입니다.')
else : 
  print('입력 오류')
```

```
#6번
# 연도가 주어졌을 때, 윤년이면 True, 아니면 False를 출력하는 프로그램을 작성하시오.
# 윤년은 연도가 4의 배수이면서, 100의 배수가 아닐때 또는 400의 배수일때이다.
# 예를 들어, 2012년은 4의 배수라서 윤년이지만, 1900년은 4의 배수이지만, 100의 배수이기때문에 윤년이 아니다.
    # 하지만 2000년은 400의 배수이기 때문에 윤년이다.


year = int(input('연도를 입력하세요'))

if year % 400 == 0 :
  print('true')
else :
  if (year % 4 == 0) and (year % 100) != 0 :
    print('true')
  else : 
    print('false')

```

```
#7번
#가위바위보 프로그램
#A와 B가 각각 가위바위보중에서 하나 낸다. 그래서 A가 이기는지 B가 이기는지 비기는지를 확인할 수 있는 프로그램을 출력해라.

import random

A_choice = random.randint(1,3)
B_choice = int(input('B는 어떤 것을 냅니까? (가위:1, 바위:2, 보:3)'))

if B_choice == 1 or B_choice == 2 or B_choice == 3 :
  if A_choice == B_choice :
    print('비겼습니다.')
  elif (A_choice == 1 and B_choice == 3) or (A_choice == 2 and B_choice == 1) or (A_choice == 3 and B_choice == 2) :
    print('A가 이겼습니다')
  else :
    print('B가 이겼습니다.')
else :  
  print('에러')
```
