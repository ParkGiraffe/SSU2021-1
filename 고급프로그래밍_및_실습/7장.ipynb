```python
#1번
#아래의 a, b, c 문자열의 문자를 조합하여 다음 조건들을 동시에 만족하는 ID 리스트를 생성하시오.
#조건1: a, b, c에서 각 문자 하나씩 뽑아서 3개의 문자를 가진 ID이어야 한다.
#조건2: ID 리스트에 중복된 ID는 없어야 한다. 
#조건3: 세트 자료구조를 사용해야한다.

a="jongj"
b="hunhu"
c="choic"

set1 = set(a)
set2 = set(b)
set3 = set(c)

result = {x+y+z for x in set1 for y in set2 for z in set3}
print(result)
print(len(result))
```

```python
#2번
#비밀번호를 만들려고 합니다.
#입력받은 비밀번호가 다음의 조건을 만족하지 않으면 만족할 때 까지 다시 입력받는 프로그램을 구현하시오.
#조건1: x가 있으면 안됩니다.
#조건2: @가 반드시 포함되어 있어야 합니다.
#조건3: 최대 9자여야 합니다.

#정답예시: whdgjs31@
#오답예시1: whdgjx31@
#오답예시2: whdgjx311
#오답예시3: whdgjs3131@



while True :
  pw = input('비밀번호를 입력하세요')

  if pw.count('x') == 0 and pw.count('@') > 0  and (len(pw) <= 9) :
    print('비밀번호가 저장되었습니다.')
    print(pw)
    break
  else : 
    print('다시 입력하세요')
    continue
```

```python
#3번
#3월 6월 9월 모의고사의 수학성적이 나와있다.
#해당 수학성적의 추이를 통해 수능 성적을 예측할 수 있는 식은 다음과 같다.
#3월,6월,9월 성적의 평균이 70점보다 높으면 그대로 나오고 70점보다 작거나 같으면 -3점을 한 결과가 예측된다.
#예를 들면 평균이 71점이면 그대로 나오고 69점이면 66점으로 예측될 것이다.
#다음 학생들의 성적이 있을 때 수능 성적예측을 하는 딕셔너리로 만들어라,
march={"피곤해":80,"고심해":48,"최준":68}
june={"피곤해":72,"고심해":72,"최준":56}
sept={"피곤해":64,"고심해":80,"최준":88}

#정답예시: november={"피곤해":72,"고심해":63,"최준":70.6}

score_dic = {
    "피곤해": [march["피곤해"], june["피곤해"], sept["피곤해"]],
    "고심해": [march["고심해"], june["고심해"], sept["고심해"]],
    "최준": [march["최준"], june["최준"], sept["최준"]]
}

november={}

for name, scores in score_dic.items():
  average = sum(scores) / len(scores)
  if average <= 70:
    average = average -3
  november[name] = average

print(november)


```

```python
#4번
#영어식 표현으로 되어있는 이름 A, B. C를 입력했을 때 한국어순으로 표현을 변환해주는 함수를 만드시오.
#변환 예시: "jong-hun-choi" => "choijonghun"

A="young-me-kim"
B="su-jin-park"
C="jae-hee-lee"

def func1(x) :
  list1 = x.split("-")
  list1.insert(0, list1[-1])
  list1.pop(-1)
  return ''.join(list1)

print(func1(A))
print(func1(B))
print(func1(C))
```

```python
#5번
#set의 내장함수를 사용해서 다음 그림의 빨간색 부분의 원소를 출력하세요.

A={1,2,3,4,5,6,7,8,9}
B={2,3,6,7,8}
C={0,2,3,5,9,11}

set1 = A.intersection(B) | B.intersection(C) | C.intersection(A)
set2 = B.intersection(C)

result = set1.difference(set2)

print(set1)
print(set2)
print(result)

```
