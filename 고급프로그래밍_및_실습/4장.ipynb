```
#4번: 1부터 100사이의 숫자 중 3의 배수의 총합을 for 문을 이용해 출력하십시오.

sum = 0

for i in range (1, 101, 1) :
  if i%3 == 0 :
     sum = sum + i


print (sum)


```

```
#5번: 다음의 리스트 요소중에서 가장 큰 값과 두번째로 큰 값을 찾아 출력하세요.
a=[10, 20, 30, 40, 5, 4 ,3, 99, 55, 100, 42, 100, 0, 105, 2, 11]

first = 0
second = 0

for i in range (0, 16) : 
  if a[i] > first :
    first = a[i]

  elif a[i] > second :
    second = a[i]

print(first, second)

    
```

```
#6번: 다음 리스트 요소중에서 70이상의 숫자는 b라는 리스트에 70미만의 숫자는 c라는 리스트에 담으세요.
a=[10, 20, 30, 40, 5, 4 ,3, 99, 55, 100, 42, 100, 0, 105, 2, 11]
b=[]
c=[]

for i in range (16) :
  if a[i] >= 70 :
    b.append(a[i])
  else :
    c.append(a[i])

print (b,c)

```

```
#7번: 교재 178쪽 숫자 퀴즈프로그램
# 숫자 맞추기(up and down)이다. 숫자를 계속 입력받아 맞출때까지 걸리는 시도횟수를 출력하는 프로그램을 만들어라
import random

tries=0
guess=0
answer=random.randint(1,100)

print("1부터 100사이의 숫자를 맞춰라")

while True:
  #여기서부터 답을 쓰세요.
  guess = int(input('답을 입력하세요. :'))
  tries = tries + 1
  if guess == answer : 
    break
  elif guess > answer : 
    print('down')
  elif guess < answer :
    print('up')
  
print ('답을 맞추셨습니다. 시도 횟수는 %s 입니다.' %(tries))
```
