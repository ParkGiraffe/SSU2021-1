```
[[6, 12, 21, 16, 1], [25, 18, 8, 4, 22], [9, 11, 13, 15, 17], [7, 24, 14, 2, 23], [10, 3, 5, 20, 19]]

```



조건:

*   단어 검색시 대소문자 구분을 하지 않아야 함 (one, One, ONE 모두 동일)




````python

def main() :

  dict1 = {}

  def display_menu () :
    print('--영한 사전 프로그램--')
    print('1. 단어 뜻 추가 \n2. 단어 검색 \n0. 프로그램 종료 \n')
    print('--------------------')
    return int(input(' > 메뉴 항목을 선택하십시오 :'))

  def func1() : 
    word = input('단어를 입력하세요')
    list1=[]
    for a in range(len(word)) : 
      list1.append(word[a].lower())
      list2 = ''.join(list1)
    return list2


  while True :
    user = display_menu()
    if user == 1 :
      word = func1()
      meaning = input('뜻을 입력하세요')
      dict1[word] = meaning

    elif user == 2 :
      word = func1()
      
      print(word,'의 뜻은', dict1[word],'입니다')
    
    elif user == 0 : 
      print('사전프로그램을 종료합니다')
      break

    else :
      print('번호를 잘못 입력하셨습니다. 다시 입력해주세요.')


main()
````

[문제 7]

Box 클래스를 작성해보자. 
최소한 다음과 같은 속성과 동작을 정의해야 한다.

Private 속성:
*  너비 
*  높이
*  깊이

동작:
*   생성자 (너비, 높이, 깊이 초기화)
*   크기 설정자 (너비, 높이, 길이 값 변경을 위함)
*   부피 획득 함수 (너비x높이x깊이) 
*   <, >, <=, >= 연산에 대한 특수 메소드 (부피 값을 비교하여 True, False 반환)
*   == 연산에 대한 특수 메소드 (너비, 높이, 깊이가 모두 같으면 True 그렇지 않으면 False 반환)
*   != 연산에 대한 특수 메소드 (너비, 높이, 깊이가 하나라도 다르면 True 그렇지 않으면 False 반환)
*   print() 함수를 이용해 객체를 출력했을 때 크기와 부피에 대한 정보를 보여줘야 함

조건:
*   모든 속성은 Private 설정 되어야 함

클래스 사용 예시:


```
A = Box()
B = Box(10, 20, 15)
    
print(A.getVolume())   #출력: 1000
print(B.getVolume())   #출력: 3000

print(A <= B)          #출력: True          
print(A < B)           #출력: True 
print(A >= B)          #출력: False  
print(A > B)           #출력: False 
print(A == B)          #출력: False 
print(A != B)          #출력: True

A.setDimension(20, 20, 40)
B.setDimension(20, 20, 40)

print(A.getVolume())   #출력: 16000
print(B.getVolume())   #출력: 16000

print(A <= B)          #출력: True
print(A < B)           #출력: False
print(A >= B)          #출력: True
print(A > B)           #출력: False
print(A == B)          #출력: True
print(A != B)          #출력: False
```












````python

class Box : 
    def __init__ (self, width = 10, height = 10, depth = 10) : 
      self.__width = width
      self.__height = height
      self.__depth = depth

    def getVolume(self) :
      return self.__width *  self.__height * self.__depth
    
    def setDimension (self, width = 10, height = 10, depth = 10) : 
      self.__width = width
      self.__height = height
      self.__depth = depth

    def __le__(self, other) : 
      return self.__width * self.__height * self.__depth <= other.__width * other.__height * other.__depth

    def __lt__(self, other) : 
      return self.__width * self.__height * self.__depth < other.__width * other.__height * other.__depth

    def __ge__(self, other) : 
      return self.__width * self.__height * self.__depth >= other.__width * other.__height * other.__depth

    def __gt__(self, other) : 
      return self.__width * self.__height * self.__depth > other.__width * other.__height * other.__depth

    def __eq__(self, other) : 
      return (self.__width == other.__width) and (self.__height == other.__height) and (self.__depth == other.__depth)

    def __neq__(self, other) : 
      return (self.__width != other.__width) or (self.__height != other.__height) or (self.__depth != other.__depth)

    def getInfo(self) : 
      print(f'너비는 {self.__width} 높이는 {self.__height} 깊이는 {self.__depth}이고, 부피는 {self.getVolume()}입니다.')


A = Box()
B = Box(10, 20, 15)
 
print(A.getVolume())   #출력: 1000
print(B.getVolume())   #출력: 3000
 
print(A <= B)          #출력: True          
print(A < B)           #출력: True 
print(A >= B)          #출력: False  
print(A > B)           #출력: False 
print(A == B)          #출력: False 
print(A != B)          #출력: True
 
A.setDimension(20, 20, 40)
B.setDimension(20, 20, 40)
 
print(A.getVolume())   #출력: 16000
print(B.getVolume())   #출력: 16000
 
print(A <= B)          #출력: True
print(A < B)           #출력: False
print(A >= B)          #출력: True
print(A > B)           #출력: False
print(A == B)          #출력: True
print(A != B)          #출력: False

A.getInfo()
B.getInfo()

print(A)
````
