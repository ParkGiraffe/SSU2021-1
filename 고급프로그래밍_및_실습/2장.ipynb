```
#2번
#다음의 숫자조합을 이용해서 피타고라스의 정리를 이용해서 직각삼각형의 빗변의 길이를 구해라
#빗변의 길이는 소수점 둘째짜리까지 구해라
#2,4 / 4,5 / 10,12 / 3,6
#math 모듈을 사용해서 문제풀이
import math
print(math.pow(25,0.5))

print(round(math.pow(2**2 + 4**2,0.5),2))
print(round(math.pow(4**2 + 5**2,0.5),2))
print(round(math.pow(10**2 + 12**2,0.5),2))
print(round(math.pow(3**2 + 6**2,0.5),2))
```

```
#3번
#다음 글자중에 잘못된 문자를 알맞게 고치세요.(힌트: 대입 & replace)
text="""
안녕하세요. 저는 19살 숭실초등학교에 다니고 있는 고급이입니다. 저는 프로그래밍을 공부하고 있는 개발자로서 파이톤을 잘 사용합니다.
또한, 볼펜으로 쓴 글을 지우개로 잘 지웁니다. 박지성은 야구선수로 한국 최초 프리미어리거입니다.
밤에 선글라스를 쓰면 앞이 잘 보입니다. 무한도전은 강호동이 진행하는 프로그램이었습니다.
"""

text = text.replace("초등학교","고등학교")
text = text.replace("파이톤","파이썬")
text = text.replace("볼펜","연필")
text = text.replace("야구선수","축구선수")
text = text.replace("잘 보입니다","안 보입니다")
text = text.replace("강호동","유재석")

print(text)

```
