```
#1번 소수판별함수를 만들고 숫자를 입력받아 소수인지 아닌지 출력하라


def prime(number) : 
  if number != 1 :
    for i in range (2, number) :

      if number % i ==0:
        return False
  else :
    return False
  return True

a = int(input('판별하고 싶은 수를 입력하세요'))

if prime(a) :
  print (a, '는 소수가 맞습니다.')
else : 
  print (a, '는 소수가 아닙니다.')
```

```
#3번 두 개의 정수를 입력받아서 최대공약수와 최소공배수를 출력하는 함수를 만들자.

def gcd (x,y) :
  if x<y :
    min =x
  else:
    min =y
  for i in range(min+1, 1, -1) :
    if x%i == 0 and y%i ==0:
      result = i
      break
  return result

# 유클리드 호제법
def abc(x,y) :
  while y :  
    x,y = y, x%y
  return x

def acd(x,y) :
  return x*y // abc(x,y)


a = int(input('첫번째 수를 입력하세요'))
b = int(input('두번째 수를 입력하세요'))

print('최대 공약수는', gcd(a,b))
print('최소 공배수는', acd(a,b))
```

```
#최대공약수 최소공배수 구하는 방식

a=int(input("첫 번째 정수: "))
b=int(input("두 번째 정수: "))

def gcd(x, y):
  if x<y:
    min=x
  else:
    min=y
  for i in range(min+1, 1, -1):
    if x%i==0 and y%i==0:
      res=i
      break
  return res

def lcm(x, y):
  if x>y:
    max=x
  else:
    max=y
  for i in range(max, (x*y)+1):
    if i%x==0 and i%y==0:
      res=i
      break
  return res

print("최대공약수는: %d" %(gcd(a, b)))
print("최소공배수는: %d" %(lcm(a, b)))
```

```
#6번 인간과 컴퓨터가 주사위게임을 하는 프로그램을 작성해보자!
#출력 예시:
#========================= 게임 호출 ========================
#인간: 주사위값=6
#컴퓨터: 주사위값 =2
#인간승리
#========================= 게임 복귀 =========================

#게임 끝낼까요?? Y/N 


import random as r

def dice_play():
    while True:
        print("========================= 게임 호출 ========================")
        a = r.randint(1, 6)
        b = r.randint(1, 6)
        print(f"인간: 주사위 값 = {a}\n컴퓨터: 주사위 값 = {b}")
        if a > b:
            print("인간 승리")
        elif a < b:
            print("컴퓨터 승리")
        else:
            print("무승부")
        print("========================= 게임 복귀 ========================\n")
        while True:
            cmd = input("게임 끝낼까요? y/n")
            if cmd == 'y':
                return
            elif cmd == 'n':
                break
            else:
                continue

dice_play()

```
