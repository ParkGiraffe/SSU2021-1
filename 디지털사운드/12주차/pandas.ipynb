**삭제**

```python
df.drop(3)
#행 삭제
```

```python
df.drop([1,4, 5, 7])
```

```python
df.drop('Class', axis=1, inplace=False)

#열 삭제
#axis = 1은 열의 방향, 0은 행의 방향
# df = df.drop을 해야 저장이 됨. 안 하면 df로 불러왔을 때 초반에 설정한 테이블이 나옴.
# 이때, inplace = Ture로 할 경우, 바로 저장이 됨.
```

```python
df #위에서 inplace True로 할 경우 열이 사라진 것이 반영됨.
```

**그룹화**

```python
df_grp = df.groupby('Class').count() #count(), sample(), mean(), max(), median(), sum()
#count : 갯수
#min : 해당 클래스의 파일 중 가장 작은 id를 출력
#max : min의 반대

#df_grp = df.groupby('Class').count()/5435*100 #비율로 표시

df_grp
```

```python
df['numeric_class'] = df['Class'].astype('category').cat.codes
#class를 카테고리화 해서 numeric_class라는 열을 추가하고 저장.
#astype : data type을 바꿈.

df
```
