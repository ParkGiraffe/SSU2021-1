```python
import librosa
import librosa.display
import IPython.display as ipd
#음성 파일을 재생할 수 있는 인터페이스를 만들어주는 라이브러리.
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
colab_path='/content/gdrive'
file_path = colab_path+'/MyDrive/Colab Notebooks/디지털사운드/9주차/02_AdjustAmplitude.wav'
y, sr = librosa.load(file_path, sr=48000, offset = 10, duration=5) #sr(sampling rate): default 22050 원본의 샘플링 레이트랑 맞춰야 함.
#sr : 샘플링레이트, 1초에 48000번 진동. 주파수.
#y : numpy.array type의 값이 있음. 샘플의 개수. 음악파일 자체라고 봐도 될 듯.
#Offset : 값만큼 앞 부분을 잘라냄.
#duration : 잘리고 남은 부분 중 지정된 값만큼 출력.
duration=librosa.get_duration(y, sr)
#duration : 음성파일의 길이 출력. (y,sr)을 해주어야 원래 값이 나온다. sr이 없으면 default 값으로 해석해서 길이가 이상하게 나옴.
len(y) #number of samples, 샘플의 개수.
len(y)/sr #이렇게 해도 duration이 나옴. 전체 샘플 개수 / 초당샘플 개수이기 때문.
```

```python
fig = plt.figure(figsize=(10,4))
librosa.display.waveplot(y, sr=sr, alpha=0.4, color='#0000ff')
#alpha : 투명도
#waveform으로 보여주는 함수
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
ipd.Audio(y, rate=sr)
#재생버튼
```
