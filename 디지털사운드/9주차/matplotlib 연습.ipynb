```python
import matplotlib.pyplot as plt
import numpy as np

#matplotlib : 그림그리기에 효과적
# 일반적으로 음악 툴 제작에서는 이 둘을 같이 씀. 그리고 pit,np라고 불러옴.
```

```python
data = {'apple':10, 'orange': 20, 'lemon': 5, 'lime': 15}
names = list(data.keys())
values = list(data.values())
names = np.array(names)
values = np.array(values)
```

```python
plt.bar(names, values)
#리스트와 array 둘 다 똑같은 그래프가 나옴.
```

```python
#subplots api : pit.subplots(값의 개수, 출력되는 그림의 개수, 사이즈 , sharey : 각각의 y값을 통일해서 볼 것인가 여부)
#fig : 그림 하나, ax : 그림 개수 (그림이 늘어나면 arry 값으로 들어옴. 따라서 ax[n]으로 써야 함.)
fig, ax = plt.subplots(1, 3, figsize=(9,3), sharey=True)
ax[0].plot(names, values)
ax[1].scatter(names, values)
ax[2].bar(names, values)
```
